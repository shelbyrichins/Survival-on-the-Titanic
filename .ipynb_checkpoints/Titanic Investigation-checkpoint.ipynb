# Investigating the Data

### What percent of all people in the dataset survived?

```python
print("Percent of passengers that survived:")
100.0 * titanic_df['Survived'].sum()/titanic_df['Survived'].count()
```

### What factors made people more likely to survive? 

#### Hypotheses: 
- __Pclass__: 1st class passengers had the highest survival, followed by 2nd class passengers
- __Fare__: Passengers who paid greater fares were more likely to survive
- __Sex__: Female passengers were more likely to survive than male passengers
- __Age__: Passengers under 18 years of age were more likely to survive
- __Ticket__: Passengers were more likely to survive if they were traveling in groups
- __SibSp__ and __Parch__: Passengers were more likely to survive if they were traveling with family

## Class

#### First, let's group the passengers by class, then examine survival in each class: 

```python
# group by class
class_group = titanic_df.groupby('Pclass')

# Examine the number of people in each passenger class: 
class_group['PassengerId'].count()
```

```python
# create plot of survival by class
%pylab inline

class_survival = class_group['Survived'].mean().plot(kind = 'bar', title = "Survival by Class", rot = 1, ylim = [0,1])
```

```python
# Calculate the proportion of passengers in each class that survived: 
class_group['Survived'].mean()
```

As expected, the highest survival belonged to 1st class, followed by 2nd class. 3rd class passengers were the least likely to survive. 
Are these differences significant? To test signficance levels, I will use a chi-squre test of independence. 

#### Test of signficance

```python
# Create a contingency table 
class_contingency = pd.crosstab(titanic_df['Survived'], titanic_df['Pclass'])
class_contingency
```

```python
# Do a chi-square test of independence to test for equality of proportions between populations:
import scipy.stats

scipy.stats.chi2_contingency(class_contingency)
```

The above values, in order of appearance, are: 
- Χ² value
- p-value 
- degrees of freedom
- array of expected values

The sample suggests there is a highly signficant difference between survival rates of 1st, 2nd, and 3rd class passengers (P = 4.5e-23)

## Fare
__Fare__ and __Pclass__ are most likely highly correlated. Let's first examine this relationship: 

```python
# examine a histogram of fare for each class

import matplotlib.pyplot as plt

def class_fare_plot(df, Pclass):
    class_fares = df[df['Pclass'] == Pclass]['Fare']
    plt.hist(class_fares, label = 'Class: {}'.format(Pclass))

class_fare_plot(titanic_df, 1)
class_fare_plot(titanic_df, 2)
class_fare_plot(titanic_df, 3)

# add a legend
plt.legend(loc = 'upper right')
```

```python
# Examine the fares of people in each passenger class: 
class_group['Fare'].describe()
```

As expected, 1st class passengers paid higher fares on average than 2nd and 3rd class passengers. 2nd class passengers paid fares more similar to 3rd than 2st class passengers. As we have already seen a relationship between class and survival, we should expect to see the same relationship with __Fare__.

__Out of curiosity, who paid fares of zero dollars? __

```python
titanic_df[titanic_df['Fare'] == 0]
```

```python
titanic_df[titanic_df['Fare'] == 0].describe()
```

Among those that did not pay any fare, most were 2nd class or higher, and all were males with no close family members on board. Additionally, all boarded at Southamption (S). I wonder if these are invited guests, or if they are perhaps higher level employees? 

Also, it is notable that only 1 (6.7%) of the passengers that rode for free survived. 

#### Examine the relationship between __Fare__ and survival:

```python
# divide data into survivors and non-survivors. We will use these data frames later as well when we examine other variables. 
survivors = titanic_df[titanic_df['Survived'] == 1]
non_survivors = titanic_df[titanic_df['Survived'] == 0]

# create histograms of Fares paid by survivors and non-survivors: 
plt.hist(survivors['Fare'], label = 'Survivors', alpha = 0.7, bins = 20)
plt.hist(non_survivors['Fare'], label = 'Non-Survivors', alpha = 0.7, bins = 20)
plt.legend(loc = 'upper right')
plt.title('Fares Paid by Survivors and Non-Survivors')
```

```python
# Descriptive stats on fares paid by survivors and non-survivors
titanic_df.groupby('Survived')['Fare'].describe()
```

Fares paid by survivors tended to be a higher than fares paid by non-survivors. 

Would like to do a statistical test. A t-test is not appropriate because the data is highly skewed, and there are some large outliers fare among those that survived.   

#### Mann-Whitney Rank Sum Test
To compare fares paid by survivors and non-survivors

```python
# Use mannwhitneyu test from the scipy stats package
scipy.stats.mannwhitneyu(survivors['Fare'], non_survivors['Fare'], alternative = 'greater')
```

A signficant difference (P = 2.3e-22) was found between the fares paid by survivors and non-survivors. Higher fares were associated with higher survival chances.

#### Let's look at survival by sex in each class: 

```python
def plot_survival_by_sex(df, Pclass):
    subset = df[df['Pclass'] == Pclass]
    subset.groupby('Sex')['Survived'].mean().plot(kind = 'bar', title = "Class {}: Survival by Sex".format(Pclass), rot = 1, ylim = [0,1])
    print(subset.groupby('Sex')['Survived'].mean())

# First class 
plot_survival_by_sex(titanic_df, 1)
```

```python
# Second class
plot_survival_by_sex(titanic_df, 2)
```

```python
# 3rd class
plot_survival_by_sex(titanic_df, 3)
```

The difference in survival between males and females was highest in 2nd class, with 92% of females surviving and only 16% of males surviving. 

1st class males were twice as likely to survive than 2nd class males. In 1st class, 97% of females and 37% of males survived. 

In  3rd class, survival was more similar between males and females, with 50% of females surviving and 14% of males surviving. Relative to 1st and 2nd class females, females in 3rd class were only half as likely to survive. 

This evidence supports the hypothesis that female survival was prioritized. It also reinforces the previous finding that 3rd class passengers had limited chances of survival. 

## Age

Were children prioritized for survival? Are the patterns consistent accross all classes and both sexes? 

First, let's examine the distribution of ages: 

```python
# What was the distribution of ages? 
titanic_df['Age'].describe()
```

Median age was 28 years old. About 25% of passengers were 20 years old or younger. 

It is also necessary to examine how age varied among the classes and sexes, because these are potentially confounding variables. 

__How did ages vary between classes?__

```python
# Because there are some NaNs in the Age column, we need to drop the rows with NaNs before plotting.
# first remove rows where age is NaN
titanic_knownAge = titanic_df.dropna(subset=['Age'], how='all')

class1_ages = titanic_knownAge[titanic_knownAge['Pclass'] == 1]['Age']
class2_ages = titanic_knownAge[titanic_knownAge['Pclass'] == 2]['Age']
class3_ages = titanic_knownAge[titanic_knownAge['Pclass'] == 3]['Age']

# Plot a histogram of the different age classes for each passenger class: 
plt.hist(class3_ages, label = '3rd class', alpha = 0.9)
plt.hist(class2_ages, label = '2nd class', alpha = 0.9)
plt.hist(class1_ages, label = '1st class', alpha = 0.9)

plt.legend(loc = 'upper right')
```

3rd class passengers tended to be a younger. Very few 1st class passengers were children.

__Were there differences in age between the genders?__

```python
# separate males from females, and drop any rows where age is NaN
female_ages = titanic_knownAge[titanic_knownAge['Pclass'] == 1]['Age']
male_ages = titanic_knownAge[titanic_knownAge['Pclass'] == 2]['Age']

# Create histograms of age classes for males and females: 
plt.hist(female_ages, label = 'Females', alpha = 0.9)
plt.hist(male_ages, label = 'Males', alpha = 0.9)

plt.legend(loc = 'upper right')
```

Males tended to be a bit younger in age. Young children (<10 years) tended to be male.

__How did survival differ between the age classes?__

```python
# Plot a histogram of ages for survivors and non-survivors: 
plt.hist(survivors['Age'].dropna(), label = 'Survivors', alpha = 0.7, bins = 20)
plt.hist(non_survivors['Age'].dropna(), label = 'Non-Survivors', alpha = 0.7, bins = 20)
plt.legend(loc = 'upper right')
plt.title('Ages of Survivors and Non-Survivors')
```

Children were a bit more likely to survive than adults. Adults aged 20-30 were the least likely to survive. Past age 30, the distributions are similar in age between survivors and non-survivors. 

__Let's calculate survival of children under the age of 18, and adults 18 years and older:__

```python
# Colculate the propotion of adults and children that survived:
children = titanic_knownAge[titanic_knownAge['Age'] < 18]
adults = titanic_knownAge[titanic_knownAge['Age'] >= 18]

print("Child (< 18 years) survival:")
print(children['Survived'].mean())

print("Adult (>= 18 years) survival:")
print(adults['Survived'].mean())
```

#### Is the difference in survival between children and adults signficant? 

```python
# Create a contingency table for a chi-square test
isChild_contingency = pd.crosstab(titanic_knownAge['Survived'], titanic_knownAge['Age'] < 18)
isChild_contingency
```

```python
# Do a chi-square test of independence to tst for equality of proportions between populations:
scipy.stats.chi2_contingency(isChild_contingency)
```

The above values, in order of appearance, are: 
- Χ² value
- p-value 
- degrees of freedom
- array of expected values

The sample suggests there is a signficant difference between survival rates adults and children (P = 0.0023)

__Did sex affect survival of children?__

```python
# Create barplot to compare the mean survival of male and female children
children.groupby('Sex')['Survived'].mean().plot(kind = 'bar', title = "Children: Survival by Sex", rot = 1, ylim = [0,1])
```

```python
# Create barplot to compare the mean survival of male and female adults
adults.groupby('Sex')['Survived'].mean().plot(kind = 'bar', title = "Adults: Survival by Sex", rot = 1, ylim = [0,1])
```

Sex did appear to affect survival by children, however survival was more similar between the sexes for children than for adults.

## Ticket
Is individal survival independent of the survival of those closest to them? Do people who travel with others survive more often? 

__First, let's look at the distribution of number of people per ticket:__

```python
# Create a dataframe of tickets
ticket_group = titanic_df[['Survived', 'Ticket']].groupby('Ticket')

# Calculate the proportion of people on each ticket that survived: 
ticket_df = ticket_group.sum()
ticket_df['Count'] = ticket_group.count()
ticket_df['Pct_Survival'] = ticket_df['Survived']/ticket_df['Count']

# Create histogram of the number of people on each ticket
ticket_df['Count'].hist()
```

Most tickets only had one person on them.

__How did survival vary by number of people on a ticket?__

```python
# Aggregate tickets over the number of people on them
ticket_df_group = ticket_df.groupby('Count')
ticket_df_group[['Pct_Survival']].describe()
```

It appears that the highest percent survival occurred for tickets with 2-3 people on them. However, there is very sparse data for tickets with more than 4 people on them. 

Perhaps we can gain more insight by looking at people who traveled with family members.

## Close Family
Siblings, spouses, parents, children.

Does traveling with family increase survival? Does this extend across classes? Across genders? 

First examine numbers of parents and children in each class group:

```python
print("Numbers of parents and children on board for travelers in each passenger class:")
class_group['Parch'].describe()
```

The number of parents and children of travelers did not vary much by passenger class

```python
print("Numbers of spouses and sibling on board for travelers in each passenger class:")
class_group['SibSp'].describe()
```

Passengers in third class were a bit more likely to travel with a spouse or sibling(s), but not by a large amount. 

Does survival change when traveling with family members? 

```python
def family_plot(survivors, non_survivors, survivor_column, non_survivor_column, plot_title):
    plt.hist(non_survivor_column, label = 'Non-Survivors', alpha = 0.7, bins = 20)
    plt.hist(survivor_column, label = 'Survivors', alpha = 0.7, bins = 20)
    plt.legend(loc = 'upper right')
    plt.title(plot_title)

# compare number of parents and children aboard for survivors and non-survivors: 
family_plot(survivors, non_survivors, survivors['Parch'], non_survivors['Parch'], "Numbers of Parents and Children Aboard")
```

```python
# compare number of siblings and spouses aboard for survivors and non-survivors: 
family_plot(survivors, non_survivors, survivors['SibSp'], non_survivors['SibSp'], "Numbers of Siblings and Spouses Aboard")
```

The distributions appear very similar for both spouses/siblings and parents/children. Let's therefore calculate the total number of close family members aboard: 

```python
# compare number of family members aboard for survivors and non-survivors: 
# First calculate the total number of close family members aboard: 
survivor_family = survivors['Parch'] + survivors['SibSp']
non_survivor_family = non_survivors['Parch'] + non_survivors['SibSp']

family_plot(survivors, non_survivors, survivor_family, non_survivor_family, "Numbers of Close Family Members Aboard")
```

Travelers with no close family members aboard are more likely to belong to the non-survivor group. 

```python
# Create a contingency table for a chi-square test
family_contingency = pd.crosstab(titanic_df['Survived'], titanic_df['Parch'] + titanic_df['SibSp'])
family_contingency
```

```python
# Do a chi-square test of independence to tst for equality of proportions between populations:
scipy.stats.chi2_contingency(family_contingency)
```

The above values, in order of appearance, are: 
- Χ² value
- p-value 
- degrees of freedom
- array of expected values

#### The sample suggests there is a signficant difference between survival rates of travelers with different numbers of close family members aboard (P = 3.6e-14)

```python
# Calculate the survival rates of passengers with different numbers of close family members aboard
print("Proportion of travelers that survived versus the number of family members they were traveling with:")
family_survival = pd.crosstab(titanic_df['Survived'], titanic_df['Parch'] + titanic_df['SibSp'], normalize='columns')
family_survival.iloc[[1]]
```

We see the same pattern as we saw in number of people per ticket: Survival peaked when traveling with 2-3 family members. 

__Does this pattern differ between males and females?__ 

```python
# Look at survival pattern amongst females
print("Female passengers")
family_survival_females = pd.crosstab(titanic_df[titanic_df['Sex'] == 'female']['Survived'], titanic_df[titanic_df['Sex'] == 'female']['Parch'] + titanic_df['SibSp'], normalize='columns')
family_survival_females.iloc[[1]]
```

```python
# Look at survival pattern amongst males 
print("Male passengers")
family_survival_males = pd.crosstab(titanic_df[titanic_df['Sex'] == 'male']['Survived'], titanic_df[titanic_df['Sex'] == 'male']['Parch'] + titanic_df['SibSp'], normalize='columns')
family_survival_males.iloc[[1]]
```

Females had high survival regardless of the number of family members they traveled with, except in the 4 - 10 family member range, where data was scarce. __Males were more than twice as likely to survive if they traveled with 2 or more family members than if they traveled alone.__ Only 16% of males that traveled alone survived, while 50% of males that traveled with three family members survived.
